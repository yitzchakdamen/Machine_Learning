```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import word2number as w2n
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(3)
```

```python
df.rename(columns={"per capita income (US$)":"per_capita_income"}, inplace=True)
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df.per_capita_income, marker=".", color="red")
```

```python
reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(df[["year"]], df[["per_capita_income"]])
```

```python
reg.predict([[2020]])
```

```python
reg.predict(df[["year"]])
```

```python
new_df = df.copy()
new_df.drop(columns=["per_capita_income"], inplace=True)
new_df.to_csv("new_data.csv", index=False)
```

```python
data = pd.read_csv("new_data.csv")

plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df.per_capita_income, marker=".", color="red")
plt.plot(data.year, reg.predict(data[["year"]]))
```
